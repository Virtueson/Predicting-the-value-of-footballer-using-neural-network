```python
import pandas as pd
```

```python
standard = pd.read_csv('top5_stats_standard.csv')
defense = pd.read_csv('top5_stats_defense.csv')
gca = pd.read_csv('top5_stats_gca.csv')
passing = pd.read_csv('top5_stats_passing.csv')
pass_type = pd.read_csv('top5_stats_passing_types.csv')
shooting = pd.read_csv('top5_stats_shooting.csv')
possession = pd.read_csv('top5_stats_possession.csv')
keeper = pd.read_csv('top5_stats_keeper.csv')
table = pd.read_csv('top5_table.csv')
```

```python
def new_id_column(dataframe):
    dataframe['ID']  = dataframe['Player'] + dataframe['Season_End_Year'].astype(str)
    dataframe.drop_duplicates(subset='ID',keep= False, inplace=True)
```

```python
dataframe_list = [standard,defense,gca,passing,pass_type,shooting,possession]
```

```python
for dataframe in dataframe_list:
    new_id_column(dataframe)
```

# Coba gabungin 

```python
conditions = ["ID"]

# Perform the multi-condition inner joins
footballer = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(standard, defense, on=conditions), gca, on=conditions), passing, on=conditions), pass_type, on=conditions), shooting, on=conditions), possession, on=conditions)
```

```python
footballer
```

```python
footballer_no_duplicates_columns = footballer.loc[:, ~footballer.columns.duplicated()]
```

```python
footballer_no_duplicates_columns
```

```python
footballer_no_duplicates_columns['Squad_Year'] = footballer_no_duplicates_columns['Season_End_Year_x'].astype(str)+footballer_no_duplicates_columns['Squad_x'] 
keeper['Squad_Year'] = keeper['Season_End_Year'].astype(str)+keeper['Squad'] 
```

```python
table['Squad_Year'] = table['Season_End_Year'].astype(str)+table['Squad']
```

```python
table = table[['Squad_Year','Rk']]
table
```

```python
conditions2 = ["Squad_Year"]
```

```python
footballer_with_rank = pd.merge(footballer_no_duplicates_columns,table,on = conditions2)
keeper_with_rank = pd.merge(keeper,table,on = conditions2)
```

```python
footballer_with_rank['Rk'].unique()
```

```python
footballer_with_rank.to_csv("footballer_stats_from_R.csv", index = False)
```
